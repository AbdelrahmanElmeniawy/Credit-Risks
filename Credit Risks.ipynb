```python
import pandas as pd
data = pd.read_csv("credit-g.csv")
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data.iloc[:, 1:2] = scaler.fit_transform(data.iloc[:, 1:2]) #duration
data.iloc[:, 4:5] = scaler.fit_transform(data.iloc[:, 4:5]) #credit_amount
data.iloc[:, 12:13] = scaler.fit_transform(data.iloc[:, 12:13]) #age
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in range(21):
    x = data.iloc[:, i:i+1]
    if x.loc[0].dtype != 'O':
        continue
    data.iloc[:, i:i+1] = encoder.fit_transform(x)
```

```python
data
```

```python
cor = data.corr()['class']
cor.sort_values(ascending = False)
```

```python
y = data['class']
X = data[['duration','credit_amount','credit_history','foreign_worker','installment_commitment','purpose','personal_status','age','savings_status','checking_status']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
X_test, X_val, y_test, y_val = train_test_split(X_train, y_train, test_size=0.5, random_state=1) 

for k in range (1, 16):
    print('k =', k)
    model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    score = model.score(X_test, y_test)
    print('\tscore ',score)
    y_pred=model.predict(X_test)
    print('\taccuracy score:', accuracy_score(y_test, y_pred))
    y_pred = model.predict(X_test)
    conf = confusion_matrix(y_test, y_pred)
    print(conf)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = 'lbfgs', multi_class = 'auto')
model.fit(X_train, y_train)
y_pred = model.predict(X_test) 
score = model.score(X_test,y_test)
from sklearn.metrics import precision_score
print('Score:', score)
print('precision score:',precision_score(y_test, y_pred, average='weighted'))
```
